```python
import json 
from tqdm import tqdm 
review = json.load(open(r'C:\Users\hp\OneDrive\Desktop\Projects\place-recommendation\travel_recommender\ds\attractionReviews.json', 'rb')) + json.load(open(r'C:\Users\hp\OneDrive\Desktop\Projects\place-recommendation\travel_recommender\ds\hotelReviews.json', 'rb')) + json.load(open(r'C:\Users\hp\OneDrive\Desktop\Projects\place-recommendation\travel_recommender\ds\ShowUserReview.json', 'rb')) 

dfs = []
for view in tqdm(review):
    
    try:
        n = len(view['reviews']) 
        df = pd.DataFrame(
            {
                'place':[view['place']]*n, 
                'reviews':view['reviews'], 
                'rating':view['rating']
            }
        )
    except Exception as e: 
        continue 

    dfs.append(df)
```

```python
from fuzzywuzzy import fuzz 

review_place = list(reviews.place.unique()) 
places_place = list(places.place.unique())

def pmatch(word, matches):
    best_score = 0
    best_match = None 
    for match in matches: 
        match_score = fuzz.ratio(word.lower().strip(), match.lower().strip())

        if match_score > best_score:
            best_score = match_score  
            best_match = match 

    return best_match

matches = {}
for word in review_place: 
    matches[word] = pmatch(word, places_place)
```

```python
reviews['place'] =  reviews.place.map(matches)
```

```python
df = pd.merge(
    left = reviews, 
    right = places, 
    on = 'place', 
    how = 'inner'
)
```

```python
df.to_csv('./ds/reviews.csv')
```
